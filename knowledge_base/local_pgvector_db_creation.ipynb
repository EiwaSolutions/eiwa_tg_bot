```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36'
}
response = requests.get('https://e-lactation.ru/sitemap2.xml', headers=headers)
sitemap_xml = response.text

soup = BeautifulSoup(sitemap_xml, 'xml')
urls = [loc.text for loc in soup.find_all('loc')]
print(urls)
```

```python
import os
import re

DIV_CLASS_PATTERN = re.compile(r'^karty-.*')

def scrape_and_write_to_txt(url):
    response = requests.get(url, headers=headers)
    page_soup = BeautifulSoup(response.text, 'html.parser')
    h1_tag = page_soup.find('h1')
    if h1_tag:
        h1_text = h1_tag.text
    else:
        print(f'No h1 tag found on {url}')
        return
    # Find the div with class 'karty-ur'
    div_tag = page_soup.find('div', class_=DIV_CLASS_PATTERN)
    if div_tag:
        div_content = div_tag.get_text(separator='\n', strip=True)
    else:
        print(f'No div with class "karty-ur" found on {url}. Trying to get div next to h1')
        div_tag = h1_tag.find_next('div')
        div_content = div_tag.get_text(separator='\n', strip=True)
        return
    filename_base = url.rsplit('/', 1)[-1]
    filename = f'{filename_base}.txt'

    with open(filename, 'w', encoding='utf-8') as file:
        file.write(f'{h1_text}\n\n{div_content}')
    print(f'Content written to {filename}')
    
```

```python
import time

for url in urls[1:]:
    scrape_and_write_to_txt(url)
    time.sleep(1)
```
